```python
from sklearnex import patch_sklearn
patch_sklearn(global_patch=True)
from autots import AutoTS
import pandas as pd
```

```python
df = pd.read_csv('don_20240502_NbArrivees_vf.csv', sep=';', parse_dates=['dhreinscription'])
df = df.rename(columns={"dhreinscription": "ds", "Total": "y"})
df = df[['ds', 'y']]
df['series_id'] = "C3"
df.head()
```

```python
model = AutoTS(
    forecast_length=7,
    frequency='D',
    prediction_interval=0.9,
    ensemble=False,
    model_list="default",  # "superfast", "default", "fast_parallel"
    transformer_list="all",  # "superfast",
    max_generations=15,
    num_validations=5,
    validation_method="backwards",
  no_negatives= True,
  n_jobs='auto',
  holiday_country='CA'
)
```

```python
model = model.fit(
    df,
    date_col='ds',
    value_col='y',
    id_col='series_id',
)
```
